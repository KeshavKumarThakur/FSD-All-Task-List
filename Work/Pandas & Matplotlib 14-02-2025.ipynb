### Introduction to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
plt.rcParams['figure.figsize'] = 8,4
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
# Distributions
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.displot(stats['InternetUsers'])
```

```python
## Box Plot 
vis2 = sns.boxplot(data= stats,x ='IncomeGroup',y='BirthRate')
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate' ,fit_reg = False ) # lm 
```

```python
vis4 = sn.lmplot(data=stats,x='InternetUsers',y = 'BirthRate')
```

```python
vis5 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup')
```

```python
vis = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',fit_reg = False,hue = 'IncomeGroup', height = 6)
```
