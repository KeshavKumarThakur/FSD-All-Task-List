```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set_style('darkgrid')

# Create subplots
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# KDE plots
k1 = sns.kdeplot(x=movies.MudgetMillions, y=movies.AudienceRating, ax=axes[0, 0])
k2 = sns.kdeplot(x=movies.MudgetMillions, y=movies.CriticRating, ax=axes[0, 1])

# Set x-axis limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Violin plot for Drama movies
z = sns.violinplot(data=movies[movies.Genre == 'Drama'], x='Year', y='CriticRating', ax=axes[1, 0])

# KDE plot with correct parameters
k4 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, fill=True, cmap='Reds', ax=axes[1, 1])

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style with custom background color
sns.set_style('dark', {'axes.facecolor': 'black'})

# Create a 2x2 subplot grid
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Plot [0,0] - KDE Plot with Different Color Maps
k1 = sns.kdeplot(x=movies.MudgetMillions, y=movies.AudienceRating, 
                 fill=True, cmap='inferno', ax=axes[0, 0])  # Fixed fill and cmap
k1b = sns.kdeplot(x=movies.MudgetMillions, y=movies.AudienceRating, 
                  cmap='cool', ax=axes[0, 0])

# Plot [0,1] - KDE Plot for CriticRating
k2 = sns.kdeplot(x=movies.MudgetMillions, y=movies.CriticRating, 
                 fill=True, cmap='inferno', ax=axes[0, 1])
k2b = sns.kdeplot(x=movies.MudgetMillions, y=movies.CriticRating, 
                  cmap='cool', ax=axes[0, 1])

# Plot [1,0] - Violin Plot for Drama Genre
z = sns.violinplot(data=movies[movies.Genre == 'Drama'], 
                   x='Year', y='CriticRating', ax=axes[1, 0])

# Plot [1,1] - KDE Plot with Two Different Color Maps
k4 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, 
                 fill=True, cmap='Blues_r', ax=axes[1, 1])
k4b = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, 
                  cmap='gist_gray_r', ax=axes[1, 1])

# Set x-axis limits for KDE plots
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Display the plots
plt.show()
```
