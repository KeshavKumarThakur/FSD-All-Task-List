### State machine Interface 

```python
x = np.linspace(0,2,100)
plt.plot(x,x, label='linear')
plt.plot(x, x**2 , label = 'quadractic')

plt.plot(x,x**3 , label ='cubic')
plt.xlabel('this is x axes')
plt.ylabel('This is Y axes')

plt.title('Simple Plot ')
plt.legend()

plt.show()
```

```python
# Error Bar Chart 
```

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

### Stacked Bar Chart 

```python
a = [15., 30., 45., 22.]
b = [15.,25.,50.,20.]
z2 = range(4)
plt.bar(z2,a,color='b')
plt.bar(z2,b,color='r',bottom=a)
plt.show()
```

### adding a legend 

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow']);
```
