# Boxplot or Whisperplot 
Box plot was was first introduced in year 1969 by Mathematician John Tukey.Box plot give a statical summary of the features being plotted.
Top line represent the max value,top edge of box is third Quartile, middle edge represents the median,bottom edge represents the first quartile value.
The bottom most line respresent the minimum value of the feature.The height of the box is called as Interquartile range.
The black dots on the plot represent the outlier values in the data.

```python
fig=plt.gcf()
fig.set_size-inches(10,7)
fig=sns.boxplot(x='Species',,y='PetalLengthCm',data=iris,order=['Iris-virginica','Iris-versicolor','Iris-setosa'],linewidth=2.5,orient='v',dodge=False)
```

```python
# Load seaborn's built-in iris dataset
iris = sns.load_dataset('iris')

# Create the figure
fig = plt.gcf()
fig.set_size_inches(10, 7)

# Create the boxplot
sns.boxplot(x='species', y='petal_length', data=iris, 
            order=['virginica', 'versicolor', 'setosa'], 
            linewidth=2.5, dodge=False)

# Show the plot
plt.show()
```

##  Violin Plot 
It is used to visualize the distribution of data and its probability distribution.
This chart is a combination of a Box Plot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data. 
The thick black bar in the centre represents the interquartile range, the thin black line extended from it represents the 95% confidence intervals, and the white dot is the median.Box Plots are limited in their display of the data, as their visual simplicity tends to hide significant details about how values in the data are distributed

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig=sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=iris)
```

##    Heatmap
Heat map is used to find out the correlation between different features in the dataset.High positive or negative value shows that the features have high correlation.This helps us to select the parmeters for machine learning.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets

# Load iris dataset
iris_data = datasets.load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

# Create the figure and set size
fig = plt.gcf()
fig.set_size_inches(10, 7)

# Plot heatmap
sns.heatmap(
    iris.corr(), 
    annot=True, 
    cmap='cubehelix', 
    linewidths=1, 
    linecolor='k', 
    square=True, 
    mask=False, 
    vmin=-1, 
    vmax=1, 
    cbar_kws={'orientation': "vertical"}, 
    cbar=True
)

# Show the plot
plt.show()
```

#  Kde Plot Kernel Desnity Estimatations 
A Kernel Density Estimate (KDE) Plot is used to visualize the distribution of a dataset smoothly, similar to a histogram but without discrete bins.

```python
sub = iris[iris['Species'] == 'setosa']  # 'setosa' instead of 'Iris-setosa'

# Create KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=sub["SepalLengthCm"], 
    y=sub["SepalWidthCm"], 
    fill=True, 
    cmap="plasma", 
    thresh=0, 
    levels=50  # Adjust levels for better contour density
)

# Titles and labels
plt.title('Iris-setosa KDE Plot')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')

# Show plot
plt.show()
```

```python
# Stacked Histogram 
```

```python
# Initialize lists
list1 = []
mylabels = []

# Loop through unique species
for gen in iris["Species"].unique():
    list1.append(iris[iris["Species"] == gen]["SepalLengthCm"])
    mylabels.append(gen)

# Plot stacked histogram
plt.figure(figsize=(10, 6))
plt.hist(list1, bins=30, stacked=True, rwidth=0.9, label=mylabels)

# Add legend and show
plt.legend()
plt.xlabel("SepalLengthCm)")
plt.ylabel("Count")
plt.title("Stacked Histogram of Sepal Length by Species")
plt.show()
```

With Stacked Histogram we can see the distribution of Sepal Length of Different Species together.This shows us the range of Sepan Length for the three different Species of Iris Flower.
